```python
import pandas as pd
import re
```

# Dataset News

```python
df = pd.read_csv('news_sentiment_analysis.csv')
df
```

```python
data_test = df[['Description', 'Sentiment']]
data_test
```

```python
def clean_comments_yt(text):
    # Hapus mention, hashtag, dan link
    text = re.sub(r'@[A-Za-z0-9_]+', '', text)
    text = re.sub(r'#\w+', '', text)
    text = re.sub(r'https?://\S+', '', text)
    text = re.sub(r'a href', '', text)

    # Hapus emoji (unicode emoticon)
    emoji_pattern = re.compile(
        "["
        u"\U0001F600-\U0001F64F"  # emoticon wajah
        u"\U0001F300-\U0001F5FF"  # simbol & pictogram
        u"\U0001F680-\U0001F6FF"  # transportasi & simbol
        u"\U0001F1E0-\U0001F1FF"  # bendera (kode negara)
        u"\U00002700-\U000027BF"  # simbol tambahan
        u"\U000024C2-\U0001F251"  # simbol enclosed
        "]+", flags=re.UNICODE)
    text = emoji_pattern.sub(r'', text)

    # Hapus karakter non-alfanumerik dan ekstra spasi
    text = re.sub(r'[^A-Za-z0-9 ]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()

    return text

data_test['Description'] = data_test['Description'].apply(clean_comments_yt)
```

```python
data_test['Description'].isna().sum()
```

```python
print("Statistik deskriptif kolom 'textDisplay':")
print(data_test['Description'].describe())
```

```python
data_test.duplicated(subset='Description').sum()
```

```python
data_test = data_test.drop_duplicates(subset='Description')
data_test.shape
```

```python
data_test.to_csv('data_test.csv')
```

```python
data_train = data_test[['Description']].copy()
data_train
```

```python
data_train.rename(columns={'Description': 'comments'}, inplace=True)
```

```python
data_train.reset_index(drop=True, inplace=True)
```

```python
data_train.to_csv('data_train.csv')
```
